```python
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

digits = load_digits()
x = digits.data
y = digits.target
```

```python
from sklearn.model_selection import  KFold

kf = KFold(n_splits=2)
```

```python
for train_index, test_index in kf.split([1,2,3,4,5,6,7,8,9]):
  print(train_index, test_index )
```

```python
def get_socre(model, x_train, x_test, y_train, y_test):
  model.fit(x_train, y_train)
  return model.score(x_test, y_test)
```

```python
from sklearn.model_selection import StratifiedKFold

fold = StratifiedKFold(n_splits=2)
```

```python
scores_l = []
scores_svm = []
scores_rf = []

for train_index, test_index in kf.split(digits.data):
  x_train, x_test, y_train, y_test = digits.data[train_index], digits.data[test_index], digits.target[train_index], digits.target[test_index]
  scores_l.append(get_socre(LogisticRegression(max_iter=200), x_train, x_test, y_train, y_test))
  scores_svm.append(get_socre(SVC(), x_train, x_test, y_train, y_test))
  scores_rf.append(get_socre(RandomForestClassifier(), x_train, x_test, y_train, y_test))
```

```python
print("RandomForestClassifier scores: ",scores_rf)
print("SVM scores: ",scores_svm)
print("LogisticRegression scores: ",scores_l)
```

```python
from sklearn.model_selection import cross_val_score
print("LogisticRegression scores: ",cross_val_score(RandomForestClassifier(), digits.data, digits.target))
print("SVM scores: ",cross_val_score(SVC(), digits.data, digits.target))
print("RandomForestClassifier scores: ",cross_val_score(LogisticRegression(max_iter=200), digits.data, digits.target))
```
