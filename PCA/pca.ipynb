```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import plotly.express as ax
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'Math': [80, 90, 70, 60, 85],
    'Science': [85, 95, 65, 70, 90],
    'English': [78, 88, 72, 68, 82]
})

ax.scatter_3d(x=df.Math, y=df.Science, z=df.English)
```

```python
scaler = StandardScaler()
scal_data = scaler.fit_transform(df)
print(scal_data)
```

```python
pca = PCA(n_components=2)
pcs_data = pca.fit_transform(scal_data)
print(pcs_data)
```

```python
plt.scatter(pcs_data[:,0], pcs_data[:,1])
plt.show()
```
